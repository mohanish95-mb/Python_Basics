**Module 1 Challenge: The "Daily Rate" Calculator**

**Goal**: Write a script that asks a team member for their hourly rate and how many hours they
worked this week.
1. Ask for the user's name.
2. Ask for their hourly_rate (use float() to allow decimals).
3. Ask for their hours_worked.
4. Calculate the total pay.
5. Print a message like: "Hey [Name], your total earnings for this week are $[Total]

```python
user_name = input("Enter your Name: ")
hourly_rate = float(input("Enter your hourly rate: "))
hours_worked = float(input("Enter your hours worked: "))
total_pay = hourly_rate * hours_worked
print(f"Hey {user_name}, your total earnings for this week are ${total_pay}")
```

**Module 2 Final Challenge: The "API Response Cleaner"**
**Scenario:** You pulled data from an API, but it gave you a list of dictionaries with some redundant info.
Python

raw_data =
[
  {"id": 1, "status": "active", "tag": "AI"},
  {"id": 2, "status": "inactive", "tag": "Cloud"},
  {"id": 3, "status": "active", "tag": "AI"}
]

Task:
1. Use a loop and .items() to print the id and tag of every "active" user.
2. Create a Set called unique_tags and add all the tags from the list to it.
3. Print the unique_tags to see how many distinct categories you have.

```python
####################### USE .items() ###########################################
raw_data = [
 {"id": 1, "status": "active", "tag": "AI"},
 {"id": 2, "status": "inactive", "tag": "Cloud"},
 {"id": 3, "status": "active", "tag": "AI"}
]

for i in raw_data:                                   # 'i' will iterate through each dictionary in the List
  for key,value in i.items():                                # 'j' acts as a tuple. It iterates through each key,value pair in the 'i.items()'. It holds (key,value), example: j[0]=key,j[1]=value
    if key == "status" and value == "active":        # If key is "status" and value is "active" then print respective ID and Tag.
      print(i["id"], i["tag"])

########################## CREATE A SET AND PRINT TAGS #########################
unique_tags = set()
for i in raw_data:
  unique_tags.add(i["tag"])
print(unique_tags)
```

**Module 3 Challenge: The "Threshold Filter"**
**Scenario:** You have a list of user engagement scores. You need to identify which
users are "Power Users" (score > 80) and store their scores in a new list.
Python

scores = [45, 88, 92, 33, 79, 100, 67]

Task:
1. Create an empty list called power_users.
2. Use a for loop to check each score.
3. If the score is above 80, .append() it to the power_users list.
4. Bonus: Try to do the same task using a List Comprehension in just one line.
5. Print the result: "Identified [X] power users with scores: [List]

```python
scores = [45, 88, 92, 33, 79, 100, 67]
power_users = []
for i in scores:
  if i > 80:
    power_users.append(i)
print(power_users)
power_users_comp = [i for i in scores if i > 80]         # List Comprehension
print(f"Identified {len(power_users)} power users with scores: {power_users_comp}")

```

**Module 4 Hands-on: The "Data Summary Tool"**

**Goal:** Create a reusable function that takes a list of numbers and returns a dictionary with the "High" and "Low" values.
**Task:**
1. Define a function called get_stats(data_list).
2. Inside the function, use Python's built-in max() and min() functions.
3. Return a dictionary that looks like this: {"max": X, "min": Y}.

```python
salaries = [50000, 120000, 85000, 200000, 45000]
max_min = dict()
def get_stats(salaries):
    max_value = max(salaries)
    min_value = min(salaries)
    return {"max":max_value, "min":min_value}
max_min = get_stats(salaries)
print(max_min)
```
